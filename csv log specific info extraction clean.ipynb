### Importing pandas library to read csv file

```python
import pandas as pd
data = pd.read_csv('mongo.csv')
data
```

```python
data['log'][10]
```

### Extracting and saving log column to a list 

```python
list1=[]
for column in data['log']:
    list1.append(column)
list1
```

### Importing regular expression library to find particular info from the item in a list

```python
import re
```

```python
for m in list1:
    for s in re.findall(r'-?\d+\.?\d*', m[-7:]):
        print(s)
```

### Saving the log greater than protocol above 500ms into a new list

```python
l=[]
for m in list1:
    for s in re.findall(r'-?\d+\.?\d*', m[-7:]):
        #print(s)
        if int(s)>=500:
            #print(m)
            l.append(m)
```

```python
len(l)
```

### Saving the output to a file

```python
with open('output1.txt', 'w') as f:
    for item in l:
        f.write("%s\n" % item)
```
